```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
name_to_id_df = pd.read_csv('species codes.csv', encoding='unicode_escape')
name_to_id_dict = {}
for _, row in name_to_id_df.iterrows():
    name_to_id_dict["{}- {}".format(row['species_primary_common_name'], row['species_scientific_name'])] = row['species_id']
    name_to_id_dict["{}-{}".format(row['species_primary_common_name'], row['species_scientific_name'])] = row['species_id']
```

```python
reference_fnames = ['pvt_andaman_and_nicobar_islands.csv',
 'pvt_rajasthan.csv',
 'pvt_arunachal_pradesh.csv',
 'pvt_madhya_pradesh.csv',
 'pvt_assam.csv',
 'pvt_jharkhand.csv',
 'pvt_maharastra.csv',
 'pvt_karnataka.csv',
 'pvt_jammu_and_kashmir.csv',
 'pvt_pvt_haryana.csv',
 'pvt_manipur.csv',
 'pvt_delhi.csv',
 'pvt_tripura.csv',
 'pvt_odisha.csv',
 'pvt_tamil_nadu.csv',
 'pvt_gujarat.csv',
 'pvt_andhra_pradesh.csv',
 'pvt_bihar.csv',
 'pvt_kerala.csv',
 'pvt_puducherry.csv',
 'pvt_meghalaya.csv']

citizen_fnames = ['Andaman and Nicobar Islands_citizenData.csv',
                  'Rajasthan_citizenData.csv',
                  'Arunachal Pradesh_citizenData.csv',
                  'Madhya Pradesh_citizenData.csv',
                  'Assam_citizenData.csv',
                  'Jharkhand_citizenData.csv',
                  'Maharashtra_citizenData.csv',
                  'Karnataka_citizenData.csv',
                  'Jammu and Kashmir_citizenData.csv',
                  'Haryana_citizenData.csv',
                  'Manipur_citizenData.csv',
                  'Delhi_citizenData.csv',
                  'Tripura_citizenData.csv',
                  'Odisha_citizenData.csv',
                  'Tamil Nadu_citizenData.csv',
                  'Gujarat_citizenData.csv',
                  'Andhra Pradesh_citizenData.csv',
                  'Bihar_citizenData.csv',
                  'Kerala_citizenData.csv',
                  'Puducherry_citizenData.csv',
                  'Meghalaya_citizenData.csv']

cat_attributes = ['Leaves_fresh', 'Leaves_mature', 'Leaves_old', 'Flowers_bud', 'Flowers_open', 'Flowers_male', 'Flowers_Female', 'Fruits_unripe', 'Fruits_ripe', 'Fruits_open']
val_df = pd.DataFrame(columns=['State Name', 'proportion of -2, -1, 0 <-> NA', 'Number of -2, -1, 0 <-> NA', 'Number of {1, 2} <-> NA', 'pct of total'])

for reff, citf in list(zip(reference_fnames, citizen_fnames)):
    print("calculating proportions for {}".format(reff))
    df_cit = pd.read_csv(os.path.join("citizenData", citf))
    df_ref = pd.read_csv(os.path.join("reference_data", reff))

    species_idweek_to_data_dict = {}
    for _, row in df_ref.iterrows():
        data = list(map(lambda attr: row[attr], cat_attributes))
        species_idweek_to_data_dict[(row['week'], row['species_id'])] = data

    not_match_count = 0
    match_count = 0
    for _, row_i in df_cit.iterrows():
        week = row_i['Week']
        try:
            species_id = name_to_id_dict[row_i['Species_name']]
        except:
            # print("Warning: formatting issue with {}".format(row_i['Species_name']))
            pass
        try:
            ref_row = species_idweek_to_data_dict[(row_i['Week'], species_id)]
        except:
            continue
        for attr in cat_attributes:
            ref_row_attr_val = ref_row[cat_attributes.index(attr)]
            if ref_row_attr_val == -2:
                if row_i[attr] == -2 or row_i[attr] == -1 or row_i[attr] == 0:
                    match_count += 1
                else:
                    not_match_count += 1
    
    val_df.loc[len(val_df)] = {'State Name': reff.strip('pvt_'), 'proportion of -2, -1, 0 <-> NA': match_count/(match_count+not_match_count), 'Number of -2, -1, 0 <-> NA': match_count, 'Number of {1, 2} <-> NA': not_match_count, 'pct of total': (match_count + not_match_count)/(10 * len(df_cit))}
    
```
