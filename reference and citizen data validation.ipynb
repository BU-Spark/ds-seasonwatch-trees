```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
name_to_id_df = pd.read_csv('species codes.csv', encoding='unicode_escape')
name_to_id_dict = {}
for _, row in name_to_id_df.iterrows():
    name_to_id_dict["{}- {}".format(row['species_primary_common_name'], row['species_scientific_name']).lower()] = row['species_id']
    name_to_id_dict["{}-{}".format(row['species_primary_common_name'], row['species_scientific_name']).lower()] = row['species_id']
print(name_to_id_dict)
```

```python
df_bihar = pd.read_csv("citizenData/Andhra Pradesh_citizenData.csv")
ref_bihar = pd.read_csv('reference_data/pvt_andhra_pradesh.csv')

cat_attributes = ['Leaves_fresh', 'Leaves_mature', 'Leaves_old', 'Flowers_bud', 'Flowers_open', 'Flowers_male', 'Flowers_Female', 'Fruits_unripe', 'Fruits_ripe', 'Fruits_open']

species_idweek_to_data_dict = {}
for _, row in ref_bihar.iterrows():
    data = list(map(lambda attr: row[attr], cat_attributes))
    species_idweek_to_data_dict[(row['week'], row['species_id'])] = data

count = 0
for _, row_i in df_bihar.iterrows():
    week = row_i['Week']
    try:
        species_id = name_to_id_dict[row_i['Species_name']]
    except:
        print("Warning: formatting issue with {}".format(row_i['Species_name'].lower()))
    
    for attr in cat_attributes:
        if row_i[attr] == -2 and row_i[attr] != species_idweek_to_data_dict[(row_i['Week'], species_id)]:
            ref_row=species_idweek_to_data_dict[(row_i['Week'], species_id)]
            #print("row: {}\n ref: {}\n".format(row_i, ref_row))
            count += 1

print(count/(10 * len(df_bihar)))
```
