```python

# This file contains data for all states in India
india_data = pd.read_csv('/Users/cecilywang/Documents/GitHub/ds-seasonwatch-trees/code/Fall 2024 Code/INDIA_top_30_species.csv')

#  Define phenophase columns for analysis
phenophase_columns = ['Valid Observations (Accuracy)', 'False Positive Rate', 'False Negative Rate']

# Filter valid observations across all states
valid_data_india = india_data.copy()
for column in phenophase_columns:
    valid_data_india = valid_data_india[valid_data_india[column].notnull()]

# average phenophase values by phenophase and state
avg_phenophase_india = valid_data_india.groupby(['Phenophase', 'State_name'])[phenophase_columns].mean().reset_index()

# count total observations for each Phenophase by State
total_observations_india = valid_data_india.groupby(['Phenophase', 'State_name'])['Total Observations'].sum().reset_index()

# merge average values and total counts into a single dataset for all states
final_india_df = pd.merge(avg_phenophase_india, total_observations_india, on=['Phenophase', 'State_name'])
```

```python

# Focus on Year, State_name, Phenophase, and Total Observations
phenophase_columns = ['Year', 'State_name', 'Phenophase', 'Total Observations']
phenophase_data = india_data[phenophase_columns].dropna()

#Analyze phenophase shifts over years for each state
# Aggregate total observations by Year, State, and Phenophase
phenophase_trends = phenophase_data.groupby(['Year', 'State_name', 'Phenophase'])['Total Observations'].sum().reset_index()

# Compute year-over-year changes in observations for each phenophase in each state
phenophase_trends['Observation_Change'] = phenophase_trends.groupby(['State_name', 'Phenophase'])['Total Observations'].diff()

map_visualization_data = phenophase_trends.copy()

# summary statistics 
state_phenophase_summary = map_visualization_data.groupby(['State_name', 'Phenophase']).agg({
    'Total Observations': 'sum',
    'Observation_Change': 'mean'
}).reset_index()
state_phenophase_summary.rename(columns={
    'Total Observations': 'Total_Observations_All_Years',
    'Observation_Change': 'Average_Yearly_Change'
}, inplace=True)

# Merge 
map_visualization_data = pd.merge(
    map_visualization_data,
    state_phenophase_summary,
    on=['State_name', 'Phenophase']
)
```

```python

columns_of_interest = [
    'Year', 'State_name', 'Phenophase', 'Total Observations',
    'Valid Observations (Accuracy)', 'False Positive Rate', 'False Negative Rate'
]
india_data_filtered = india_data[columns_of_interest].dropna()

# Separate data into historical reference (pre-2020) and comparative analysis (post-2020)
pre_2020_data = india_data_filtered[india_data_filtered['Year'] < 2020]
post_2020_data = india_data_filtered[india_data_filtered['Year'] >= 2020]

# Analyze phenophase shifts
# Aggregating data by State, Phenophase, and Year for both datasets
pre_2020_summary = pre_2020_data.groupby(['State_name', 'Phenophase']).agg({
    'Total Observations': 'sum',
    'Valid Observations (Accuracy)': 'mean',
    'False Positive Rate': 'mean',
    'False Negative Rate': 'mean'
}).reset_index()

post_2020_summary = post_2020_data.groupby(['State_name', 'Phenophase']).agg({
    'Total Observations': 'sum',
    'Valid Observations (Accuracy)': 'mean',
    'False Positive Rate': 'mean',
    'False Negative Rate': 'mean'
}).reset_index()

# Combine historical and comparative data
comparison_summary = pd.merge(
    pre_2020_summary, post_2020_summary,
    on=['State_name', 'Phenophase'],
    suffixes=('_pre_2020', '_post_2020')
)

#  Calculate differences to identify phenophase shifts
comparison_summary['Observation_Change'] = (
    comparison_summary['Total Observations_post_2020'] - 
    comparison_summary['Total Observations_pre_2020']
)

comparison_summary['Accuracy_Change'] = (
    comparison_summary['Valid Observations (Accuracy)_post_2020'] - 
    comparison_summary['Valid Observations (Accuracy)_pre_2020']
)

comparison_summary['False_Positive_Change'] = (
    comparison_summary['False Positive Rate_post_2020'] - 
    comparison_summary['False Positive Rate_pre_2020']
)

comparison_summary['False_Negative_Change'] = (
    comparison_summary['False Negative Rate_post_2020'] - 
    comparison_summary['False Negative Rate_pre_2020']
)
```

```python
# Load the top 30 species data (Kerala dataset)
top_30_data_path = '/Users/cecilywang/Documents/GitHub/ds-seasonwatch-trees/code/Fall 2024 Code/top_30_species_all_data.csv'
top_30_data = pd.read_csv(top_30_data_path)

#Filter for valid phenophase observations (Value 0 in `[Phenophase]_incorrect_-2` columns)
# Identify phenophase columns indicating incorrect -2 values
phenophase_columns = [col for col in top_30_data.columns if '_incorrect_-2' in col]

# Keep rows where all phenophase columns indicate valid observations (Value = 0)
valid_data = top_30_data.copy()
for column in phenophase_columns:
    valid_data = valid_data[valid_data[column] == 0]

# Separate data into pre-2020 (historical reference) and post-2020 (comparative analysis)
pre_2020_data = valid_data[valid_data['Year'] < 2020]
post_2020_data = valid_data[valid_data['Year'] >= 2020]

# Analyze valid observations by location (Lat, Long) and phenophase
# Aggregate data for pre-2020
pre_2020_summary = pre_2020_data.groupby(['Lat', 'Long', 'Species_name', 'State_name']).agg({
    'Leaves_fresh': 'mean',
    'Flowers_bud': 'mean',
    'Fruits_ripe': 'mean'
}).reset_index()
pre_2020_summary.rename(columns={
    'Leaves_fresh': 'Avg_Leaves_fresh_pre_2020',
    'Flowers_bud': 'Avg_Flowers_bud_pre_2020',
    'Fruits_ripe': 'Avg_Fruits_ripe_pre_2020'
}, inplace=True)

# Aggregate data for post-2020
post_2020_summary = post_2020_data.groupby(['Lat', 'Long', 'Species_name', 'State_name']).agg({
    'Leaves_fresh': 'mean',
    'Flowers_bud': 'mean',
    'Fruits_ripe': 'mean'
}).reset_index()
post_2020_summary.rename(columns={
    'Leaves_fresh': 'Avg_Leaves_fresh_post_2020',
    'Flowers_bud': 'Avg_Flowers_bud_post_2020',
    'Fruits_ripe': 'Avg_Fruits_ripe_post_2020'
}, inplace=True)

# Merge pre-2020 and post-2020 summaries
comparison_summary = pd.merge(
    pre_2020_summary, post_2020_summary,
    on=['Lat', 'Long', 'Species_name', 'State_name'],
    how='outer'
)

# Calculate changes in phenophase averages
comparison_summary['Change_Leaves_fresh'] = (
    comparison_summary['Avg_Leaves_fresh_post_2020'] - 
    comparison_summary['Avg_Leaves_fresh_pre_2020']
)
comparison_summary['Change_Flowers_bud'] = (
    comparison_summary['Avg_Flowers_bud_post_2020'] - 
    comparison_summary['Avg_Flowers_bud_pre_2020']
)
comparison_summary['Change_Fruits_ripe'] = (
    comparison_summary['Avg_Fruits_ripe_post_2020'] - 
    comparison_summary['Avg_Fruits_ripe_pre_2020']
)
```

```python

output_path_map_analysis = "/Users/cecilywang/Documents/GitHub/ds-seasonwatch-trees/code/Fall 2024 Code/POWERBIkeralaseasonwatch_map_analysis_top_30.csv"
comparison_summary.to_csv(output_path_map_analysis, index=False)

output_path_map_analysis
```

```python


# Only include rows where all `[Phenophase]_incorrect_-2` columns are valid (0)
phenophase_columns = [col for col in top_30_data.columns if '_incorrect_-2' in col]

valid_data = top_30_data.copy()
for column in phenophase_columns:
    valid_data = valid_data[valid_data[column] == 0]

# Add a "Pre/Post 2020" column to differentiate periods
valid_data['Period'] = valid_data['Year'].apply(lambda x: 'Pre-2020' if x < 2020 else 'Post-2020')

# Calculate weekly shifts in phenophase metrics
# Group data by Week, Period, Lat, Long, and Phenophase columns
weekly_shift_data = valid_data.groupby(
    ['Week', 'Period', 'Lat', 'Long', 'Species_name', 'State_name']
).agg({
    'Leaves_fresh': 'mean',
    'Flowers_bud': 'mean',
    'Fruits_ripe': 'mean'
}).reset_index()

# Create a pivot for week-over-week changes
# Separate the pre-2020 and post-2020 periods
pre_2020_weekly = weekly_shift_data[weekly_shift_data['Period'] == 'Pre-2020']
post_2020_weekly = weekly_shift_data[weekly_shift_data['Period'] == 'Post-2020']

# Merge the pre and post weekly averages for comparison
weekly_comparison = pd.merge(
    pre_2020_weekly, post_2020_weekly,
    on=['Week', 'Lat', 'Long', 'Species_name', 'State_name'],
    suffixes=('_pre_2020', '_post_2020')
)

# Calculate weekly changes in phenophase averages
weekly_comparison['Change_Leaves_fresh'] = (
    weekly_comparison['Leaves_fresh_post_2020'] - 
    weekly_comparison['Leaves_fresh_pre_2020']
)
weekly_comparison['Change_Flowers_bud'] = (
    weekly_comparison['Flowers_bud_post_2020'] - 
    weekly_comparison['Flowers_bud_pre_2020']
)
weekly_comparison['Change_Fruits_ripe'] = (
    weekly_comparison['Fruits_ripe_post_2020'] - 
    weekly_comparison['Fruits_ripe_pre_2020']
)
```

# Flourish and PowerBI Ideas

### **1. Map Visualization: Geographic Distribution**
- **Purpose**: Display the spatial trends in phenophase changes across locations.
- **Visual Type**: Map (with Latitude and Longitude fields).
- **Fields**:
  - **Location**: Use `Lat` and `Long` for precise points.
  - **Legend**: `Species_name` to differentiate species.
  - **Color Coding**: Use `Change_Leaves_fresh`, `Change_Flowers_bud`, or `Change_Fruits_ripe` to visualize shifts in phenophase metrics (positive = greener/blue, negative = red/orange).
  - **Tooltip**: Include:
    - `State_name`
    - `Leaves_fresh_pre_2020` vs. `Leaves_fresh_post_2020` (and similarly for other phenophases).
    - Weekly averages and changes.

---

### **2. Line Chart: Weekly Trends**
- **Purpose**: Show phenophase behavior over weeks for pre- and post-2020 periods.
- **Visual Type**: Line Chart.
- **Fields**:
  - **Axis**: `Week`.
  - **Legend**: `Period` (Pre-2020 and Post-2020).
  - **Values**: Weekly averages (`Leaves_fresh`, `Flowers_bud`, `Fruits_ripe`).
- **Filters**:
  - Allow filtering by `Species_name`, `State_name`, and location (Lat/Long).

---

### **3. Bar Chart: Species-Level Changes**
- **Purpose**: Compare average phenophase shifts by species.
- **Visual Type**: Clustered Bar Chart.
- **Fields**:
  - **Axis**: `Species_name`.
  - **Values**:
    - `Change_Leaves_fresh`, `Change_Flowers_bud`, `Change_Fruits_ripe` (use multiple bars per species for comparison).
- **Legend**: `Phenophase` to group by phenophase type.

---

### **4. Table: Detailed Data**
- **Purpose**: Provide a detailed breakdown of changes for reference.
- **Visual Type**: Table.
- **Fields**:
  - Include:
    - `Species_name`
    - `Lat`, `Long`
    - `State_name`
    - `Week`
    - `Leaves_fresh_pre_2020`, `Leaves_fresh_post_2020`, `Change_Leaves_fresh` (repeat for other phenophases).
  - Add slicers for `State_name`, `Species_name`, and `Week`.

---

### **5. Filters and Interactivity**
- **Slicers**:
  - `State_name`: To focus on specific regions.
  - `Species_name`: To drill down into species-specific trends.
  - `Phenophase`: To analyze specific phenophases (`Leaves_fresh`, `Flowers_bud`, `Fruits_ripe`).
  - `Week`: To examine particular weeks of interest.

- **Interactions**:
  - Make the map interactive with cross-filtering. Selecting a point on the map highlights corresponding line and bar charts.
  - Allow filtering on species or states to dynamically update all visuals.
